```python
%%manim -v WARNING LineGraphExampleeng
from manim import *
from math import sqrt


class LineGraphExampleeng(Scene):
    def construct(self):
        axes = Axes(
            x_range = [0, 1.25, 0.25],
            y_range = [0, 1.25, 0.25],
            x_length = 7,
            y_length=7,
        )
        axes.scale(0.8)
        axes.to_edge(LEFT)

        x_label_1 = Tex("1").next_to(axes.c2p(1, 0, 0), DOWN)
        y_label_1 = Tex("1").next_to(axes.c2p(0, 1, 0), LEFT)

       # Dodawanie ćwiartki okręgu
        quarter_circle = Arc(start_angle=PI/2, angle=-PI/2, radius=4.5, color=BLUE)
        quarter_circle.move_to(axes.c2p(0.5,0.5,0), aligned_edge=ORIGIN)
        # Dodawanie prostej y=x
        line_y_equals_x = Line(axes.c2p(0, 0, 0), axes.c2p(1.125, 0.9, 0), color=WHITE)

        # Zaznaczanie punktu przecięcia
        intersection_point = axes.c2p(5/sqrt(41), 4/sqrt(41), 0)
        intersection_dot = Dot(intersection_point, color=RED)
        intersection_point2 = axes.c2p(1, 0, 0)
        intersection_dot2 = Dot(intersection_point2, color=RED)
        punkt = axes.c2p(5/sqrt(41), 0, 0)
        dot3 = Dot(punkt, color=RED)
        punkt2 = axes.c2p(1, 0.8, 0)
        dot4 = Dot(punkt2, color=RED)
        segment_to_x_axis = Line(intersection_point, axes.c2p(5/sqrt(41), 0, 0), color=WHITE)
        segment_to_line = Line(axes.c2p(1, 0, 0), axes.c2p(1, 0.8, 0), color=WHITE)

        # Oznaczenia dla punktów
        label_O = Tex("O").next_to(axes.c2p(0, 0, 0), DOWN + LEFT)
        label_A = Tex("A").next_to(axes.c2p(5/sqrt(41), 0, 0), DOWN)
        label_B = Tex("B").next_to(intersection_point, UP)
        label_C = Tex("C").next_to(axes.c2p(1, 0, 0), UP + RIGHT)
        label_D = Tex("D").next_to(axes.c2p(1, 0.8, 0),  RIGHT)

        # Tworzenie obszaru OAB
       # polygon_OAB = Polygon(axes.c2p(0, 0, 0), axes.c2p(sqrt(2)/2, 0, 0), intersection_point, fill_opacity=0.3, fill_color=BLUE)




        self.play(
            Write(axes),
            Write(x_label_1),
            Write(y_label_1),
            
        )
        # Tworzenie kąta
        angle = Angle(axes.x_axis, line_y_equals_x, radius=1, color=WHITE)
        # Etykieta dla kąta
        label_alfa = MathTex(r"\alpha").next_to(angle, LEFT, buff=0.000125)

        self.wait(3)
        self.play(Write(quarter_circle))
        self.wait(3)
        self.play(Write(line_y_equals_x))
        self.play(Write(intersection_dot))
        self.play(Write(intersection_dot2))
        self.wait(4)
        self.play(Write(segment_to_x_axis), Write(segment_to_line))
        self.play(Write(dot3))
        self.play(Write(dot4))
        self.play(Write(label_O), Write(label_A), Write(label_B), Write(label_C), Write(label_D))
        self.wait(5)
        # self.play(Write(polygon_OAB))
        # self.wait(1)
        line_AB = Line(
        intersection_point,  
        axes.c2p(5/sqrt(41), 0, 0),
        color=GREEN
        )

        line_OB = Line(
        axes.c2p(0, 0, 0),  
        axes.c2p(5/sqrt(41), 4/sqrt(41), 0),  
        color=RED
        )
        
        


        formula1 = MathTex(r"L_{BC} = \alpha").to_edge(RIGHT).shift(UP*3+LEFT*2)
        formula_sin = MathTex(r"\sin \alpha = \frac{|AB|}{|OB|}").move_to(formula1.get_center() + DOWN*1.5)


        self.play(Write(angle), Write(label_alfa))
        self.wait(5)
        self.play(Write(formula1), run_time=2)
        self.wait(5)
        self.play(Write(formula_sin))
        self.wait(2)
        self.play(Create(line_AB), Create(line_OB))
        formula_sin[0][5].set_color(GREEN)
        formula_sin[0][6].set_color(GREEN)
        formula_sin[0][7].set_color(GREEN)
        formula_sin[0][8].set_color(GREEN)
        formula_sin[0][10].set_color(RED)
        formula_sin[0][11].set_color(RED)
        formula_sin[0][12].set_color(RED)
        formula_sin[0][13].set_color(RED)

        label_OB_length = Tex("1").next_to(line_OB, UP, buff=0.0000000000001)
        label_OB_length.shift(DOWN * 0.6)
        self.wait(4)
        self.play(Write(label_OB_length))
        self.wait(5)

        formula_sin_transformed = MathTex(r"|AB| =\sin \alpha").move_to(formula1.get_center() + DOWN*1.5)
        self.play(Transform(formula_sin, formula_sin_transformed))

        label_AB_length = MathTex(r"\sin \alpha").next_to(line_AB, RIGHT, buff=0.005)
        self.play(Write(label_AB_length))
        line_AB.set_stroke(WHITE, 2)
        line_OB.set_stroke(WHITE, 2)
        self.wait(2)




        line_CD = Line(
        axes.c2p(1, 0.8, 0),  
        axes.c2p(1, 0, 0),
        color=GREEN
        )

        line_OC = Line(
        axes.c2p(0, 0, 0),  
        axes.c2p(1, 0, 0),  # punkt B
        color=RED 
        )
        
        


       
        formula_tan = MathTex(r"\tan \alpha = \frac{|CD|}{|OC|}").move_to(formula_sin.get_center() + DOWN*1.5)
        self.play(Write(formula_tan))
        self.wait(5)
        self.play(Create(line_CD), Create(line_OC))
        formula_tan[0][4].set_color(GREEN)
        formula_tan[0][5].set_color(GREEN)
        formula_tan[0][6].set_color(GREEN)
        formula_tan[0][7].set_color(GREEN)
        formula_tan[0][9].set_color(RED)
        formula_tan[0][10].set_color(RED)
        formula_tan[0][11].set_color(RED)
        formula_tan[0][12].set_color(RED)

        formula_tan_transformed = MathTex(r"|CD| =\tan \alpha").move_to(formula_sin.get_center() + DOWN*1.5)
        self.play(Transform(formula_tan, formula_tan_transformed))
        self.wait(6)
        line_CD.set_stroke(WHITE, 2)
        line_OC.set_stroke(WHITE, 2)
        self.wait(1)
        label_CD_length = MathTex(r"\tan \alpha").next_to(line_CD, RIGHT, buff=0.005)
        self.play(Write(label_CD_length))
        self.wait(1)


        self.play(Uncreate(formula_tan), run_time = 0.5)
        self.play(Uncreate(formula_sin), run_time =0.5)
        nowa_linijka = MathTex(r" L_{BC}= \alpha, |AB| = \sin \alpha, |CD| = \tan \alpha").to_edge(RIGHT).shift(UP*3)

        BC= Line(axes.c2p(5/sqrt(41), 4/sqrt(41), 0), axes.c2p(1, 0, 0), color=WHITE)
        self.play(Create(BC))
        self.wait(1)

        
        self.play(Transform(formula1, nowa_linijka))
        self.wait(3)
        polygon_OAB = Polygon(axes.c2p(0, 0, 0), axes.c2p(1, 0, 0), intersection_point, fill_opacity=0.3, fill_color=RED, stroke_opacity=0 )
        self.play(Write(polygon_OAB))
        self.wait(1)
        wzor1 = MathTex(r" P_{OBC} = \frac{1 \cdot \sin \alpha}{2} = \frac{\sin \alpha}{2}").move_to(nowa_linijka.get_center() + DOWN*1.5)
        self.play(Write(wzor1))
        self.wait(6)
        polygon_OAB.set_fill(opacity=0)  # Zdejmij kolor
        self.play(Write(polygon_OAB))

        self.wait(7)
        wzor2 = MathTex(r" P_{\alpha} = \frac{\alpha}{2} \cdot  r^2 = \frac{ \alpha}{2}").move_to(wzor1.get_center() + DOWN*1.5)
        self.play(Write(wzor2))
        self.wait(5)


        polygon_OCD = Polygon(axes.c2p(0, 0, 0), axes.c2p(1, 0, 0), axes.c2p(1, 0.8, 0), fill_opacity=0.3, fill_color=RED, stroke_opacity=0 )
        self.play(Write(polygon_OCD))
        self.wait(3)
        wzor3 = MathTex(r" P_{OCD} = \frac{1 \cdot \tan \alpha}{2} = \frac{\tan \alpha}{2}").move_to(wzor2.get_center() + DOWN*1.5)
        self.play(Write(wzor3))
        self.wait(2)
        polygon_OCD.set_fill(opacity=0)
        self.play(Write(polygon_OCD))
        self.wait(3)
        self.play(Uncreate(wzor3), run_time = 0.5)
        self.play(Uncreate(wzor2), run_time = 0.5)
        nowosc = MathTex(r" P_{OBC}=\frac{\sin \alpha}{2}, P_{\alpha} = \frac{ \alpha}{2}, P_{OCD} = \frac{\tan \alpha}{2}").to_edge(RIGHT).shift(UP*2)
        self.play(Transform(wzor1, nowosc))
        self.wait(2)
        nowosc2 = MathTex(r" \frac{\sin \alpha}{2} \leq  \frac{\alpha}{2} \leq \frac{\tan \alpha}{2}").move_to(nowosc.get_center() + DOWN*1.5)
        self.play(Write(nowosc2))
        self.wait(3)
        nowosc2_1 = MathTex(r" / \cdot \frac{2}{\sin \alpha}").move_to(nowosc2.get_center() + RIGHT*3)
        self.play(Write(nowosc2_1))
        self.wait(1)
        nowosc3 = MathTex(r" 1 \leq  \frac{\alpha}{\sin \alpha} \leq \frac{1}{\cos \alpha}").move_to(nowosc2.get_center() + DOWN*1.5)
        self.play(Write(nowosc3))
        self.wait(4)
        nowosc4_0 = MathTex(r" 1  " )
        nowosc4_1 = MathTex(r" \geq")
        nowosc4_2 = MathTex(r" \frac{\sin \alpha}{\alpha} ")
        nowosc4_3 = MathTex(r" \geq")
        nowosc4_4 = MathTex(r" \cos \alpha")
        group_nowosc4 = VGroup(nowosc4_0, nowosc4_1, nowosc4_2, nowosc4_3, nowosc4_4).arrange(RIGHT).move_to(nowosc2.get_center() + DOWN*1.5)
        self.play(Transform(nowosc3, group_nowosc4))
        self.wait(5)
        
        arrow_to_1 = Arrow(nowosc4_0.get_bottom() + DOWN*0.05, nowosc4_0.get_bottom() + DOWN*1.6 , color=WHITE)
        label_1 = MathTex(r"1").next_to(arrow_to_1, DOWN*2 , buff=0.1)
        alpha_to_0_for_1 = MathTex(r"\alpha \to 0").next_to(arrow_to_1, LEFT*0.005  , buff=0.01).scale(0.7)

        arrow_to_cos = Arrow(nowosc4_4.get_bottom() + DOWN*0.05 , nowosc4_4.get_bottom() + DOWN*1.6, color=WHITE)
        label_cos = MathTex(r"1").next_to(arrow_to_cos, DOWN*2 , buff=0.1)
        alpha_to_0_for_cos = MathTex(r"\alpha \to 0").next_to(arrow_to_cos, RIGHT*0.005  , buff=0.01).scale(0.7)

        arrow_to_sine_over_alpha = Arrow(nowosc4_2.get_bottom() + DOWN*0.05, nowosc4_2.get_bottom() + DOWN*1.5,  color=BLUE)
        label_sine_over_alpha = MathTex(r"1").next_to(arrow_to_sine_over_alpha, DOWN, buff=0.1)
        alpha_to_0_for_sine = MathTex(r"\alpha \to 0").next_to(arrow_to_sine_over_alpha, RIGHT, buff=0.01).scale(0.7)

        self.play(
            GrowArrow(arrow_to_1),
            Write(alpha_to_0_for_1),
            
            GrowArrow(arrow_to_cos),
            Write(alpha_to_0_for_cos)
        )
        self.wait(6)
        self.play(
            Write(label_1),
            Write(label_cos)
        )
        self.wait(6)  
        
        self.play(
            GrowArrow(arrow_to_sine_over_alpha),
            Write(alpha_to_0_for_sine)
        )
        self.play(Write(label_sine_over_alpha))
        
        self.wait(7)
```
