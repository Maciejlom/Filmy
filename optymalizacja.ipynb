```python
%%manim -v WARNING wwp
from manim import *

class wwp(Scene):
    def construct(self):
        ax = Axes(
            x_range=[-6,6, 1],
            y_range=[-6,6,1],
            x_length=6,
            y_length=6, 
            axis_config={"include_numbers": True,  "font_size": 16, 'tip_shape': StealthTip},
            x_axis_config=None,
            y_axis_config=None, tips=True).move_to(RIGHT*3)

        labels = ax.get_axis_labels(
            Tex("x").scale(0.7),
            Text("y").scale(0.45)
        )

        a, b, c = 1, 2, -1

        def qf(x):
            return a * x ** 2 + b * x + c
    
        plot = ax.plot(qf, x_range=[-4, 2, 0.01], color=GREEN)


        wierzcholek = ax.c2p(-1, -2) 
        napis_z_wierzcholkiem_nap = MathTex(r"(p,q)", font_size=36, color=GOLD).next_to(wierzcholek, DOWN)
        point = Dot(wierzcholek, color=GOLD)

        linia_wartosci = DashedLine(
            start=ax.c2p(-6.1,-2), 
            end=ax.c2p(6.1,-2), 
            color=GOLD,
            stroke_width=3,
            dash_length=0.16
        )

        linia_wartosci2 = DashedLine(
            start=ax.c2p(-6.1,7), 
            end=ax.c2p(6.1,7), 
            color=GOLD,
            stroke_width=3,
            dash_length=0.16
        )

        napis_zbior = Tex("Zbiór wartości: ", font_size=26, color=GOLD).move_to(LEFT*4+UP*3)
        zbior = MathTex(r"\langle -2; +\infty)", font_size = 26, color = GOLD).next_to(napis_zbior, RIGHT)

        napis_najmn_wartosc = Tex("Najmniejsza wartość: ", font_size=26, color=GOLD).next_to(napis_zbior, DOWN)
        najmn_wartosc = MathTex(r"-2", font_size = 26, color = GOLD).move_to(napis_najmn_wartosc, RIGHT).shift(RIGHT*0.5)

        napis_najw_wartosc = Tex("Najwieksza wartość: ", font_size=26, color=GOLD).next_to(napis_najmn_wartosc, DOWN)
        najw_wartosc = MathTex(r"+\infty", font_size = 26, color = GOLD).move_to(napis_najw_wartosc, RIGHT).shift(RIGHT*0.5)

        jednak_nie = Tex(r"brak", font_size = 26, color = GOLD).move_to(napis_najw_wartosc, RIGHT).shift(RIGHT*0.8)

        gruparamka = VGroup(
            napis_zbior,
            zbior,
            napis_najmn_wartosc,
            najmn_wartosc,
            napis_najw_wartosc,
            najw_wartosc
        )

        ramka = SurroundingRectangle(gruparamka, color=GREEN, buff=0.5)

        #druga_funkcja

        a2, b2, c2 = -1, 8, -13

        def qff(x):
            return a2 * x ** 2 + b2 * x + c2
    
        plot2 = ax.plot(qff, x_range=[1, 7, 0.01], color=BLUE)


        wierzcholek2 = ax.c2p(4, 3) 
        napis_z_wierzcholkiem_nap2 = MathTex(r"(p,q)", font_size=36, color=GOLD).next_to(wierzcholek2, DOWN)
        point2 = Dot(wierzcholek2, color=GOLD)

        linia_wartoscia = DashedLine(
            start=ax.c2p(-6.1,3), 
            end=ax.c2p(7,3), 
            color=GOLD,
            stroke_width=3,
            dash_length=0.16
        )

        linia_wartosci2a = DashedLine(
            start=ax.c2p(-6.1,-6), 
            end=ax.c2p(7,-6), 
            color=GOLD,
            stroke_width=3,
            dash_length=0.16
        )

        napis_zbiora = Tex("Zbiór wartości: ", font_size=26, color=GOLD).move_to(LEFT*4)
        zbiora = MathTex(r"(-\infty; 3 \rangle", font_size = 26, color = GOLD).next_to(napis_zbiora, RIGHT)

        napis_najmn_wartosca = Tex("Najmniejsza wartość: ", font_size=26, color=GOLD).next_to(napis_zbiora, DOWN)
        najmn_wartosca = MathTex(r"-\infty", font_size = 26, color = GOLD).move_to(napis_najmn_wartosca, RIGHT).shift(RIGHT*0.5)

        napis_najw_wartosca = Tex("Najwieksza wartość: ", font_size=26, color=GOLD).next_to(napis_najmn_wartosca, DOWN)
        najw_wartosca = MathTex(r"3", font_size = 26, color = GOLD).move_to(napis_najw_wartosca, RIGHT).shift(RIGHT*0.5)

        jednak_niea = Tex(r"brak", font_size = 26, color = GOLD).move_to(napis_najmn_wartosca, RIGHT).shift(RIGHT*0.8)


        gruparamka2 = VGroup(
            napis_zbiora,
            zbiora,
            napis_najmn_wartosca,
            najmn_wartosca,
            napis_najw_wartosca,
            najw_wartosca
        )

        ramka2 = SurroundingRectangle(gruparamka2, color=BLUE, buff=0.5)

        zbior_wartosci1 = MathTex(r"\langle q; +\infty)", font_size=26, color = GOLD).next_to(napis_zbior, RIGHT)
        zbior_wartosci2 = MathTex(r" (-\infty ; q \rangle", font_size=26, color = GOLD).next_to(napis_zbiora, RIGHT)

        nnw = MathTex(r"q", font_size = 26, color = GOLD).move_to(napis_najmn_wartosc, RIGHT).shift(RIGHT*0.5)
        nnw2 = MathTex(r"q", font_size = 26, color = GOLD).move_to(napis_najw_wartosca, RIGHT).shift(RIGHT*0.5)


        a1 = MathTex(r"a > 0", font_size=36, color=GREEN).next_to(wierzcholek, LEFT).shift(DOWN*0.3)
        a2 = MathTex(r"a < 0 ", font_size=36, color=BLUE).next_to(wierzcholek2, RIGHT).shift(UP*0.3)
        


        
#odpalanie_funkcji
        
        self.play(
                Write(ax),
                Write(labels), 
                run_time = 0.6
                )
        self.wait(1)
        self.play(Write(plot), run_time = 0.6)
        self.wait()
        self.play(
            Write(point),
            Write(napis_z_wierzcholkiem_nap),
            run_time = 0.6
        )
        self.wait()
        self.play(Write(linia_wartosci), run_time = 0.6)
        self.wait()
        self.play(Write(linia_wartosci2), run_time = 0.6)
        self.wait(2)
        self.play(Write(napis_zbior), run_time = 0.6)
        self.wait(1.5)
        self.play(Write(zbior), run_time = 0.6)
        self.wait(3)
        self.play(Write(napis_najmn_wartosc), run_time = 0.6)
        self.wait(1.5)
        self.play(Write(najmn_wartosc), run_time = 0.6)
        self.wait(1.5)
        self.play(Write(napis_najw_wartosc), run_time = 0.6)
        self.wait(1.5)
        self.play(Write(najw_wartosc), run_time = 0.6)
        self.wait()
        self.play(Transform(najw_wartosc, jednak_nie))
        self.wait(2)
        self.play(
            FadeOut(linia_wartosci),
            FadeOut(linia_wartosci2),
            Create(ramka),
            FadeOut(point),
            FadeOut(napis_z_wierzcholkiem_nap)
            )

        self.wait()

#odpalanie funkcji 2

        self.play(Write(plot2), run_time = 0.6)
        self.wait()
        self.play(
            Write(point2),
            Write(napis_z_wierzcholkiem_nap2), 
            run_time = 0.6
        )
        self.wait()
        self.play(Write(linia_wartoscia), run_time = 0.6)
        self.wait()
        self.play(Write(linia_wartosci2a), run_time = 0.6)

        self.wait()
        self.play(Write(napis_zbiora), run_time = 0.6)
        self.play(Write(zbiora), run_time = 0.6)
        self.wait()
        self.play(Write(napis_najmn_wartosca), run_time = 0.6)
        self.play(Write(najmn_wartosca), run_time = 0.6)
        self.wait()
        self.play(Write(napis_najw_wartosca), run_time = 0.6)
        self.play(Write(najw_wartosca), run_time = 0.6)
        self.play(Transform(najmn_wartosca, jednak_niea))
        self.wait(0.5)
        self.play(
            FadeOut(linia_wartoscia),
            FadeOut(linia_wartosci2a),
            Create(ramka2),
            FadeOut(point2),
            FadeOut(napis_z_wierzcholkiem_nap2)
            )
        self.wait(0.5)
        self.play(
            Transform(zbior, zbior_wartosci1),
            Transform(zbiora, zbior_wartosci2)
        )
        self.wait(1)
        self.play(
            Transform(najmn_wartosc, nnw),
            Transform(najw_wartosca, nnw2)
        )
        self.play(
            Write(a1),
            Write(a2)
        )

        self.wait(2)


    
```
